```python
def minutes_converter(time):
    new_string=str(time)
    new_string=new_string[7:]
    new_string=new_string.replace(':','')
    new_string=new_string.strip('s')
    new_string=new_string.strip(' +')
    new_string=new_string.strip(' ')
    total_minutes=int(new_string[:2])*60+int(new_string[2:4])*1+int(new_string[4:])/60
    return total_minutes
#This function was used for me to get the total_time_elapsed data into a float data type
```

```python
df['Total_time_elapsed']=df['Total_time_elapsed'].apply(minutes_converter)
```

```python
def time_category(time_placed):
    time_placed=str(time_placed)
    time_placed=int(time_placed[7:9])
    if time_placed>=5 and time_placed<12:
        return 'Morning'
    elif time_placed>=12 and time_placed<17:
        return 'Afternoon'
    elif time_placed>=17 and time_placed<=20:
        return 'Evening'
    else:
        return 'Night'
#This function helped categorize the Orders placed into either Morning, Afternoon, Evening, or Night.
#I thought it would help for later data analysis
    
```

```python
new_df['Time_category']=new_df['Order_placed_time'].apply(time_category)
```

```python
def yes_disc(number):
    if int(number)>0:
        return 'Yes'
    else:
        return 'No'
#This just signified if the order placed used a discount or not
```

```python
new_df['Used_discount']=new_df['Amount_of_discount'].apply(yes_disc)
```

```python
df_for_dummies=df_for_testing.select_dtypes(include='object')
#Here is just my way of finding the correct dummy variables and then combining one final data frame
```

```python
df_done = pd.get_dummies(df_for_dummies,drop_first=True)
```

```python
final_df = pd.concat([df_exclude,df_done],axis=1)
```

```python
final_df
```

```python
def converter_timedelta(col):
    col=col.seconds
    return col
#This was used to get the timedelta columns in a form of integers
```

```python
final_df['Order_placed_time']=final_df['Order_placed_time'].apply(converter_timedelta)
```

```python
final_df['Time_restaurant_receieved_order']=final_df['Time_restaurant_receieved_order'].apply(converter_timedelta)
```

```python
final_df['Driver_restaurant_arrival_time']=final_df['Driver_restaurant_arrival_time'].apply(converter_timedelta)
```

```python
final_df['Driver_delivery_final_time']=final_df['Driver_delivery_final_time'].apply(converter_timedelta)
```

```python
final_df
```

```python
seconds_in_day = 24*60*60

final_df['sin_order_placed_time'] = np.sin(2*np.pi*final_df.Order_placed_time/seconds_in_day)
final_df['cos_order_placed_time'] = np.cos(2*np.pi*final_df.Order_placed_time/seconds_in_day)
final_df['sin_restaurant_received_time'] = np.sin(2*np.pi*final_df.Time_restaurant_receieved_order/seconds_in_day)
final_df['cos_restaurant_received_time'] = np.cos(2*np.pi*final_df.Time_restaurant_receieved_order/seconds_in_day)
final_df['sin_driver_restaurant_arrival_time'] = np.sin(2*np.pi*final_df.Driver_restaurant_arrival_time/seconds_in_day)
final_df['cos_driver_restaurant_arrival_time'] = np.cos(2*np.pi*final_df.Driver_restaurant_arrival_time/seconds_in_day)
final_df['sin_driver_delivery_final_time'] = np.sin(2*np.pi*final_df.Driver_delivery_final_time/seconds_in_day)
final_df['cos_driver_delivery_final_time'] = np.cos(2*np.pi*final_df.Driver_delivery_final_time/seconds_in_day)
#This function was used to help model the ciruclar implications of time or else these four columns would be overvalued when it came to the model
```

```python
final_df.drop(columns=['Order_placed_time','Time_restaurant_receieved_order','Driver_restaurant_arrival_time','Driver_delivery_final_time'])
```

```python
X=final_df.drop('Total_time_elapsed',axis=1)
#Setting up model
```

```python
y=final_df['Total_time_elapsed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
#Scaled data
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1,1,5,10,50,100],
              'l1_ratio':[.1, .5, .7, .9, .95, .99, 1]}
#Setup of alpha and lasso ratios
```

```python
from sklearn.model_selection import GridSearchCV
```

```python

grid_model = GridSearchCV(estimator=base_elastic_model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5,
                          verbose=1)
```

```python
grid_model.fit(scaled_X_train,y_train)
```

```python
grid_model.best_params_
#found best parameters from grid search
```

```python
y_pred = grid_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
#testing error methods
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
#testing error methods
```
